```python
train_data = pd.read_csv("/kaggle/input/playground-series-s4e1/train.csv");
train_data.head(5)
```

```python
test_data = pd.read_csv("/kaggle/input/playground-series-s4e1/test.csv");
test_data.head(5)
```

```python
train_data['CreditScore'].corr(train_data['Exited'])
```

high credit score means low chance of exiting account 

```python
train_data.info()
```

```python
features = ['CreditScore', 'Age', 'Tenure', 'Balance', 'EstimatedSalary']
train_data[features].corr()
```

```python
from sklearn.ensemble import RandomForestRegressor 

y= train_data['Exited']
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestRegressor(n_estimators = 100, max_depth = 5, random_state = 1)
model.fit(X,y)

predictions = model.predict(X_test)
output = pd.DataFrame({'id': test_data.id, 'Exited': predictions})
output
```

```python
output.to_csv('Submission1.csv', index = False)
```
